```python
import pandas as pd
import numpy as np
```

```python
def make_float(x):
    if "만원" in x:    #뛰어쓰기가 있는 경우에는 두 개로 쪼개서 계산 ex) 1123 1/3
        number1, str1 = x.split("만원")
        number = int(number1) * 10000
    else:
        if "달러" in x: #뛰어쓰기가 없지만 /가 있는경우 ex) 1/3
            number1, str1 = x.split("달러")
            number = int(number1) * 1000
        else: #숫자만 있을때 ex 112
            number = int(x)
    return number
```

```python
doosan_hdf = pd.read_csv("doosan_hsalary.csv")
doosan_pdf = pd.read_csv("doosan_psalary.csv")
hanhwa_hdf = pd.read_csv("hanhwa_hsalary.csv")
hanhwa_pdf = pd.read_csv("hanhwa_psalary.csv")
lotte_hdf = pd.read_csv("lotte_hsalary.csv")
lotte_pdf = pd.read_csv("lotte_psalary.csv")
nexen_hdf = pd.read_csv("nexen_hsalary.csv")
nexen_pdf = pd.read_csv("nexen_psalary.csv")
samsung_hdf = pd.read_csv("samsung_hsalary.csv")
samsung_pdf = pd.read_csv("samsung_psalary.csv")
KIA_hdf = pd.read_csv("KIA_hsalary.csv")
KIA_pdf = pd.read_csv("KIA_psalary.csv")
KT_hdf = pd.read_csv("KT_hsalary.csv")
KT_pdf = pd.read_csv("KT_psalary.csv")
LG_hdf = pd.read_csv("LG_hsalary.csv")
LG_pdf = pd.read_csv("LG_psalary.csv")
NC_hdf = pd.read_csv("NC_hsalary.csv")
NC_pdf = pd.read_csv("NC_psalary.csv")
SK_hdf = pd.read_csv("SK_hsalary.csv")
SK_pdf = pd.read_csv("SK_psalary.csv")
```

```python
doosan_hdf["salary"] = doosan_hdf["salary"].apply(make_float)
doosan_pdf["salary"] = doosan_pdf["salary"].apply(make_float)
hanhwa_hdf["salary"] = hanhwa_hdf["salary"].apply(make_float)
hanhwa_pdf["salary"] = hanhwa_pdf["salary"].apply(make_float)
lotte_hdf["salary"] = lotte_hdf["salary"].apply(make_float)
lotte_pdf["salary"] = lotte_pdf["salary"].apply(make_float)
nexen_hdf["salary"] = nexen_hdf["salary"].apply(make_float)
nexen_pdf["salary"] = nexen_pdf["salary"].apply(make_float)
samsung_hdf["salary"] = samsung_hdf["salary"].apply(make_float)
samsung_pdf["salary"] = samsung_pdf["salary"].apply(make_float)
KIA_hdf["salary"] = KIA_hdf["salary"].apply(make_float)
KIA_pdf["salary"] = KIA_pdf["salary"].apply(make_float)
KT_hdf["salary"] = KT_hdf["salary"].apply(make_float)
KT_pdf["salary"] = KT_pdf["salary"].apply(make_float)
LG_hdf["salary"] = LG_hdf["salary"].apply(make_float)
LG_pdf["salary"] = LG_pdf["salary"].apply(make_float)
NC_hdf["salary"] = NC_hdf["salary"].apply(make_float)
NC_pdf["salary"] = NC_pdf["salary"].apply(make_float)
SK_hdf["salary"] = SK_hdf["salary"].apply(make_float)
SK_pdf["salary"] = SK_pdf["salary"].apply(make_float)
```

```python
doosan_df = pd.concat([doosan_hdf, doosan_pdf]).reset_index(drop=True)
hanhwa_df = pd.concat([hanhwa_hdf, hanhwa_pdf]).reset_index(drop=True)
lotte_df = pd.concat([lotte_hdf, lotte_pdf]).reset_index(drop=True)
nexen_df = pd.concat([nexen_hdf, nexen_pdf]).reset_index(drop=True)
samsung_df = pd.concat([samsung_hdf, samsung_pdf]).reset_index(drop=True)
KIA_df = pd.concat([KIA_hdf, KIA_pdf]).reset_index(drop=True)
KT_df = pd.concat([KT_hdf, KT_pdf]).reset_index(drop=True)
LG_df = pd.concat([LG_hdf, LG_pdf]).reset_index(drop=True)
NC_df = pd.concat([NC_hdf, NC_pdf]).reset_index(drop=True)
SK_df = pd.concat([SK_hdf, SK_pdf]).reset_index(drop=True)
```

```python
doosan_df = doosan_df.groupby("name", as_index=False).agg({"salary":np.sum, "p_number":np.size})
hanhwa_df = hanhwa_df.groupby("name", as_index=False).agg({"salary":np.sum, "p_number":np.size})
lotte_df = lotte_df.groupby("name", as_index=False).agg({"salary":np.sum, "p_number":np.size})
nexen_df = nexen_df.groupby("name", as_index=False).agg({"salary":np.sum, "p_number":np.size})
samsung_df = samsung_df.groupby("name", as_index=False).agg({"salary":np.sum, "p_number":np.size})
KIA_df = KIA_df.groupby("name", as_index=False).agg({"salary":np.sum, "p_number":np.size})
KT_df = KT_df.groupby("name", as_index=False).agg({"salary":np.sum, "p_number":np.size})
LG_df = LG_df.groupby("name", as_index=False).agg({"salary":np.sum, "p_number":np.size})
NC_df = NC_df.groupby("name", as_index=False).agg({"salary":np.sum, "p_number":np.size})
SK_df = SK_df.groupby("name", as_index=False).agg({"salary":np.sum, "p_number":np.size})
```

```python
doosan_df["new_salary"]=np.array(doosan_df["salary"]/doosan_df["p_number"])
hanhwa_df["new_salary"]=np.array(hanhwa_df["salary"]/hanhwa_df["p_number"])
lotte_df["new_salary"]=np.array(lotte_df["salary"]/lotte_df["p_number"])
nexen_df["new_salary"]=np.array(nexen_df["salary"]/nexen_df["p_number"])
samsung_df["new_salary"]=np.array(samsung_df["salary"]/samsung_df["p_number"])
KIA_df["new_salary"]=np.array(KIA_df["salary"]/KIA_df["p_number"])
KT_df["new_salary"]=np.array(KT_df["salary"]/KT_df["p_number"])
LG_df["new_salary"]=np.array(LG_df["salary"]/LG_df["p_number"])
NC_df["new_salary"]=np.array(NC_df["salary"]/NC_df["p_number"])
SK_df["new_salary"]=np.array(SK_df["salary"]/SK_df["p_number"])
```

```python
doosan_df["team"]="두산"
hanhwa_df["team"]="한화"
lotte_df["team"]="롯데"
nexen_df["team"]="넥센"
samsung_df["team"]="삼성"
KIA_df["team"]="KIA"
KT_df["team"]="KT"
LG_df["team"]="LG"
NC_df["team"]="NC"
SK_df["team"]="SK"
```

```python
doosan_df["salary"]=doosan_df["new_salary"]/10000
hanhwa_df["salary"]=hanhwa_df["new_salary"]/10000
lotte_df["salary"]=lotte_df["new_salary"]/10000
nexen_df["salary"]=nexen_df["new_salary"]/10000
samsung_df["salary"]=samsung_df["new_salary"]/10000
KIA_df["salary"]=KIA_df["new_salary"]/10000
KT_df["salary"]=KT_df["new_salary"]/10000
LG_df["salary"]=LG_df["new_salary"]/10000
NC_df["salary"]=NC_df["new_salary"]/10000
SK_df["salary"]=SK_df["new_salary"]/10000
```

```python
doosan_df=doosan_df[["name","team","salary"]]
hanhwa_df=hanhwa_df[["name","team","salary"]]
lotte_df=lotte_df[["name","team","salary"]]
nexen_df=nexen_df[["name","team","salary"]]
samsung_df=samsung_df[["name","team","salary"]]
KIA_df=KIA_df[["name","team","salary"]]
KT_df=KT_df[["name","team","salary"]]
LG_df=LG_df[["name","team","salary"]]
NC_df=NC_df[["name","team","salary"]]
SK_df=SK_df[["name","team","salary"]]
```
